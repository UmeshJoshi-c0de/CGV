```python
import matplotlib.pyplot as plt

def draw_circle_midpoint(x_center, y_center, radius):
    # Initial point on the circumference of the circle
    x = 0
    y = radius

    # Decision parameter
    p = 1 - radius

    # Store the points to plot
    x_points = []
    y_points = []

    # Plot the initial points in all octants
    def plot_circle_points(x_center, y_center, x, y):
        x_points.extend([
            x_center + x, x_center - x, x_center + x, x_center - x,
            x_center + y, x_center - y, x_center + y, x_center - y
        ])
        y_points.extend([
            y_center + y, y_center + y, y_center - y, y_center - y,
            y_center + x, y_center + x, y_center - x, y_center - x
        ])

    plot_circle_points(x_center, y_center, x, y)

    # Generate points for the circle using the Midpoint Circle Drawing Algorithm
    while x < y:
        x += 1
        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1
        plot_circle_points(x_center, y_center, x, y)

    # Plot the circle using matplotlib
    plt.scatter(x_points, y_points)
    plt.title(f"Midpoint Circle Drawing Algorithm (Radius = {radius})")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.show()

# Example usage
x_center, y_center = 0, 0
radius = 10
draw_circle_midpoint(x_center, y_center, radius)
```
